### **Caculate the first derivative of the curve**

```python
corona_dataset_aggregated.loc["China"].diff().plot()
```

The above plot represents the infection rates in days.

### **Find the maxmimum (daily) infection rate for China, Canada, and US**

```python
# number of cases has been recorded in one day
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
# number of cases has been recorded in one day
corona_dataset_aggregated.loc["Canada"].diff().max()
```

```python
# number of cases has been recorded in one day
corona_dataset_aggregated.loc["US"].diff().max()
```

### **Find maximum infection rate for all of the countries**

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries :
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
max_infection_rates
```

```python
corona_dataset_aggregated["max_infection_rates"] = max_infection_rates
```

```python
corona_dataset_aggregated.head(10)
```

### **Create a new dataframe with only needed column**

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rates"])
```

```python
corona_data.head(10)
```

### **Drop the useless columns**

### **Join the two dataset we have prepared**

### **Correlation matrix**

Remember that correlation values range from -1 to 1:
<li>A correlation of 1 indicates a perfect positive correlation.
<li>A correlation of -1 indicates a perfect negative correlation.
<li>A correlation of 0 indicates no linear correlation.
<br>
<br>We can use this correlation matrix to analyze relationships between variables in your dataset and identify which variables are strongly correlated or not.

```python
data.corr()
```

Luckily we have positive correlation between max_infection_rates and any other features.
